# Data Preprocessing

## Encode Target Label

```python
# Create a LabelEncoder object
le = LabelEncoder()

# Fit and transform the label encoder on y_train
y_train = le.fit_transform(y_train) 
y_test = le.transform(y_test) 
```

```python
# Get the class labels
class_labels = le.classes_
print(class_labels)
# Print the class labels and their corresponding encoded values
for i, label in enumerate(class_labels):
    print(f"Class: {label}, Encoded Label: {i}")
```

# Encode categorical features and normalize nummerical features

```python
# Identify categorical and numerical columns
categorical_columns = x_train.select_dtypes(include=['object']).columns.tolist()
numerical_columns = x_train.select_dtypes(exclude=['object']).columns.tolist()

# Define the preprocessor with OrdinalEncoder for categorical columns
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OrdinalEncoder(), categorical_columns),  # Ordinal encode categorical columns
        ('num', StandardScaler(), numerical_columns)  # Pass through numerical columns
    ]
)

# Step 1: Transform the data using the preprocessor
x_train = preprocessor.fit_transform(x_train)
x_test = preprocessor.transform(x_test)
```

## Dimensionality Reduction (PCA)

```python
from sklearn.decomposition import PCA
# Dimensionality Reduction using PCA
# Choose 2 components for 2D visualization
pca = PCA(n_components=2)

# Fit PCA on the scaled training data and transform both scaled training and test data
x_train_pca = pca.fit_transform(x_train)
x_test_pca = pca.transform(x_test)
x_test = x_test_pca.copy()

print(f"Dimensionality reduced to {pca.n_components} principal components.")
print(f"Explained variance ratio of PCA components: {pca.explained_variance_ratio_}")

# 3. Visualize dataset after dimensionality reduction
# Create a DataFrame for visualization
# Assuming 'y_train' is available and aligned with x_train
plot_df_train = pd.DataFrame(data=x_train_pca, columns=['Principal Component 1', 'Principal Component 2'])
# Add y_train to the DataFrame for coloring if it's a classification task
# Make sure y_train is also aligned by index or reset index if it was not preserved from original x_train
plot_df_train['Target'] = y_train.reset_index(drop=True) if isinstance(y_train, pd.Series) else y_train


plt.figure(figsize=(12, 8))
sns.scatterplot(
    x='Principal Component 1',
    y='Principal Component 2',
    hue='Target', # Color points by target variable if applicable
    data=plot_df_train,
    alpha=0.7,
    s=50
)
plt.title('Dataset Visualization after PCA (Training Data)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='Target Class', loc='upper right')
plt.show()

print("\nVisualization of the dataset after PCA completed.")
```

# Modeling and Experiment

```python
# Define the list of Naive Bayes models
ml_models = [
    ("BernoulliNB", BernoulliNB()),
    ("GaussianNB", GaussianNB()),
]

# Perform cross-validation for each model
for name, model in ml_models:
    print(f"{name}: ")
    try:
        model.fit(x_train, y_train)  # Use training data
        skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)  # Use StratifiedKFold
        cv_scores = cross_val_score(model, x_train, y_train, cv=skf, scoring='accuracy')  # Use accuracy for classification
        print("\tStratified K-Fold Cross-Validation:", cv_scores)
        print('-'*100)
        print(f'\tAverage Stratified K-Fold Cross-Validation: {cv_scores.mean():.4f}')
        print('-'*100)
    except Exception as e:
        print(f"\tError occurred: {str(e)}")
        print('-'*100)
```

## Hyperparameters tuning

```python
# Define the StratifiedKFold cross-validator
skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

# Objective function for BernoulliNB
def objective_bernoulli(trial):
    # Define hyperparameter search space
    alpha = trial.suggest_float('alpha', 0.0, 10.0)  # Smoothing parameter
    binarize = trial.suggest_float('binarize', 0.0, 1.0) if trial.suggest_categorical('use_binarize', [True, False]) else None
    fit_prior = trial.suggest_categorical('fit_prior', [True, False])
    
    # Create the model with suggested hyperparameters
    model = BernoulliNB(
        alpha=alpha,
        binarize=binarize,
        fit_prior=fit_prior,
        class_prior=None  # Let the model compute class priors
    )
    
    # Perform cross-validation
    cv_scores = cross_val_score(model, x_train, y_train, cv=skf, scoring='accuracy')
    return cv_scores.mean()

# Objective function for GaussianNB
def objective_gaussian(trial):
    # Define hyperparameter search space
    var_smoothing = trial.suggest_float('var_smoothing', 1e-12, 1e-5, log=True)  # Log scale for small values
    
    # Create the model with suggested hyperparameters
    model = GaussianNB(
        var_smoothing=var_smoothing,
        priors=None  # Let the model compute class priors
    )
    
    # Perform cross-validation
    cv_scores = cross_val_score(model, x_train, y_train, cv=skf, scoring='accuracy')
    return cv_scores.mean()

# Perform hyperparameter tuning for each model
ml_models = [
    ("GaussianNB", objective_gaussian),
    ("BernoulliNB", objective_bernoulli),
]

for name, objective in ml_models:
    print(f"{name}: ")
    try:
        # Create an Optuna study to maximize accuracy
        study = optuna.create_study(direction='maximize')
        study.optimize(objective, n_trials=50)  # Run 50 trials for each model
        
        # Get the best parameters and score
        best_params = study.best_params
        best_score = study.best_value
        
        # Print the results
        print(f"\tBest Hyperparameters: {best_params}")
        print(f"\tBest Cross-Validation Accuracy: {best_score:.4f}")
        
        # Re-run cross-validation with the best parameters for detailed scores
        if name == "BernoulliNB":
            best_model = BernoulliNB(
                alpha=best_params['alpha'],
                binarize=best_params['binarize'] if best_params.get('use_binarize', False) else None,
                fit_prior=best_params['fit_prior'],
                class_prior=None
            )
        else:  # GaussianNB
            best_model = GaussianNB(
                var_smoothing=best_params['var_smoothing'],
                priors=None
            )
        
        cv_scores = cross_val_score(best_model, x_train, y_train, cv=skf, scoring='accuracy')
        print("\tStratified K-Fold Cross-Validation Scores:", cv_scores)
        print('-'*100)
        print(f"\tAverage Stratified K-Fold Cross-Validation: {cv_scores.mean():.4f}")
        print('-'*100)
        
    except Exception as e:
        print(f"\tError occurred: {str(e)}")
        print('-'*100)
```

## Calibration curves

```python
lr = LogisticRegression(C=1.0)
gnb = GaussianNB()
gnb_isotonic = CalibratedClassifierCV(gnb, cv=2, method="isotonic")
gnb_sigmoid = CalibratedClassifierCV(gnb, cv=2, method="sigmoid")

clf_list = [
    (lr, "Logistic"),
    (gnb, "Naive Bayes"),
    (gnb_isotonic, "Naive Bayes + Isotonic"),
    (gnb_sigmoid, "Naive Bayes + Sigmoid"),
]
fig = plt.figure(figsize=(15, 10))
gs = GridSpec(4, 2)
colors = plt.get_cmap("Dark2")

ax_calibration_curve = fig.add_subplot(gs[:2, :2])
calibration_displays = {}
for i, (clf, name) in enumerate(clf_list):
    clf.fit(x_train, y_train)
    display = CalibrationDisplay.from_estimator(
        clf,
        x_train,
        y_train,
        n_bins=10,
        name=name,
        ax=ax_calibration_curve,
        color=colors(i),
    )
    calibration_displays[name] = display

ax_calibration_curve.grid()
ax_calibration_curve.set_title("Calibration plots (Naive Bayes)")

# Add histogram
grid_positions = [(2, 0), (2, 1), (3, 0), (3, 1)]
for i, (_, name) in enumerate(clf_list):
    row, col = grid_positions[i]
    ax = fig.add_subplot(gs[row, col])

    ax.hist(
        calibration_displays[name].y_prob,
        range=(0, 1),
        bins=10,
        label=name,
        color=colors(i),
    )
    ax.set(title=name, xlabel="Mean predicted probability", ylabel="Count")

plt.tight_layout()
plt.show()
```
